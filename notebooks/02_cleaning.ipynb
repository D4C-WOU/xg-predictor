```python
import pandas as pd
df = pd.read_csv("../data/raw/international_shots.csv")

df.head()
```

```python
df.info()
```

```python
important_columns = [
    "match_id",
    "competition_name",
    "season_name",
    "minute",
    "second",
    "period",
    "team",
    "player",
    "position",
    "location",
    "shot_end_location",
    "shot_body_part",
    "shot_technique",
    "play_pattern",
    "under_pressure",
    "shot_first_time",
    "shot_one_on_one",
    "shot_open_goal",
    "shot_deflected",
    "shot_statsbomb_xg",
    "shot_outcome"
]

shots_clean = df[important_columns].copy()

print(shots_clean.shape)

shots_clean.head()
```

```python
shots_clean.isnull().sum()
```

```python
shots_clean["x"] = shots_clean["location"].apply(
    lambda loc: loc[0] if isinstance(loc, list) else None
)

shots_clean["y"] = shots_clean["location"].apply(
    lambda loc: loc[1] if isinstance(loc, list) else None
)

shots_clean["end_x"] = shots_clean["shot_end_location"].apply(
    lambda loc: loc[0] if isinstance(loc, list) else None
)

shots_clean["end_y"] = shots_clean["shot_end_location"].apply(
    lambda loc: loc[1] if isinstance(loc, list) else None
)
```

```python
shots_clean['goal'] = shots_clean['shot_outcome'].apply(
    lambda x:1 if x == 'Goal' else 0
)
```

```python
shots_clean['goal'].value_counts()
```

```python
boolean_cols = [
    "under_pressure",
    "shot_first_time",
    "shot_one_on_one",
    "shot_open_goal",
    "shot_deflected"
]

for col in boolean_cols:
    shots_clean[col] = shots_clean[col].fillna(False)
    shots_clean[col] = shots_clean[col].astype(int)
```

```python
shots_clean = shots_clean.dropna(
    subset=['x','y']
)

shots_clean.info()
```

```python
import ast
shots_clean["shot_end_location"] = shots_clean[
    "shot_end_location"
].apply(
    lambda x: ast.literal_eval(x)
    if pd.notnull(x)
    else None
)
```

```python
shots_clean["end_x"] = shots_clean[
    "shot_end_location"
].apply(
    lambda loc: loc[0]
    if isinstance(loc, list)
    else None
)

shots_clean["end_y"] = shots_clean[
    "shot_end_location"
].apply(
    lambda loc: loc[1]
    if isinstance(loc, list)
    else None
)
```

```python
shots_clean[["shot_end_location", "end_x", "end_y"]].head()
```

```python
shots_clean.info()
```

```python
shots_clean = shots_clean.drop(
    columns=[
        "location",
        "shot_end_location"
    ]
)
# error because i already dropped the column but accidentally ran the block twice/thrice
```

```python
shots_clean = shots_clean.drop_duplicates()

print(shots_clean.shape)
```

```python
shots_clean.to_csv(
    "../data/processed/cleaned_shots.csv",
    index=False
)

print("Cleaned dataset saved successfully")
```
