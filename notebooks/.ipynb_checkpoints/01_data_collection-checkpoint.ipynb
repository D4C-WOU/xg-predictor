```python
from mplsoccer import Sbopen
import pandas as pd

parser = Sbopen()
```

```python
competitions = parser.competition()
competitions.head()
```

```python
wc = competitions[competitions["competition_name"]=="FIFA World Cup"]
wc
```

```python
matches =  parser.match(
    competition_id = wc.iloc[0]['competition_id'],
    season_id = wc.iloc[0]['season_id']
)
matches.head()
```

```python
events,related,freeze,tactics = parser.event(matches.iloc[0]['match_id'])
events,related,freeze,tactics.head()
```

```python
shots = events[events['type_name'] == 'Shot']
shots.head()
```
